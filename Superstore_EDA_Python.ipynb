## 🧹 Data Cleaning

```python
# Convert dates to datetime format
orders['Order Date'] = pd.to_datetime(
    orders['Order Date'], dayfirst=True)
orders['Ship Date'] = pd.to_datetime(
    orders['Ship Date'], dayfirst=True)

# Extract Year and Month from Order Date
orders['Year'] = orders['Order Date'].dt.year
orders['Month'] = orders['Order Date'].dt.month
orders['Month Name'] = orders['Order Date'].dt.strftime('%B')

print("Dates converted successfully !")
print("\nYears available :", sorted(orders['Year'].unique()))
```

```python
# Create Profit Margin column
orders['Profit Margin'] = orders['Profit'] / orders['Sales']

# Create Discount Group column
orders['Discount Group'] = pd.cut(
    orders['Discount'],
    bins=[-0.01, 0, 0.20, 0.40, 1.0],
    labels=['No Discount', 'Low (0-20%)',
            'Medium (20-40%)', 'High (40%+)']
)

print("New columns created successfully !")
print(orders[['Sales', 'Profit', 'Profit Margin',
              'Discount', 'Discount Group']].head())
```

## 📊 Business Overview

```python
# Global KPIs
total_sales = orders['Sales'].sum()
total_profit = orders['Profit'].sum()
total_orders = orders['Order ID'].nunique()
avg_discount = orders['Discount'].mean()
avg_margin = orders['Profit Margin'].mean()

print("=" * 40)
print("GLOBAL BUSINESS OVERVIEW")
print("=" * 40)
print(f"Total Sales      : ${total_sales:,.0f}")
print(f"Total Profit     : ${total_profit:,.0f}")
print(f"Total Orders     : {total_orders:,}")
print(f"Average Discount : {avg_discount:.1%}")
print(f"Average Margin   : {avg_margin:.1%}")
print("=" * 40)
```

## 📦 Sales Analysis by Category

```python
# Sales and Profit by Category
category = orders.groupby('Category').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum'),
    Total_Orders=('Order ID', 'count')
).reset_index()

category['Profit Margin'] = (
    category['Total_Profit'] / category['Total_Sales'] * 100
).round(2)

category = category.sort_values(
    'Total_Sales', ascending=False)

print(category.to_string(index=False))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Chart 1 - Sales by Category
axes[0].bar(category['Category'],
            category['Total_Sales'] / 1e6,
            color=['#1F3864', '#2980B9', '#95A5A6'])
axes[0].set_title('Total Sales by Category ($M)',
                   fontsize=13, fontweight='bold',
                   color='#1F3864')
axes[0].set_xlabel('')
axes[0].set_ylabel('Sales ($M)')

# Chart 2 - Profit Margin by Category
axes[1].bar(category['Category'],
            category['Profit Margin'],
            color=['#2ECC71', '#E8722A', '#E74C3C'])
axes[1].set_title('Profit Margin % by Category',
                   fontsize=13, fontweight='bold',
                   color='#1F3864')
axes[1].set_xlabel('')
axes[1].set_ylabel('Margin %')

plt.tight_layout()
plt.savefig('category_analysis.png',
            dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved !")
```

## 🌍 Regional Analysis

```python
# Sales and Profit by Region
region = orders.groupby('Region').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum'),
    Total_Orders=('Order ID', 'count')
).reset_index()

region['Profit Margin'] = (
    region['Total_Profit'] / region['Total_Sales'] * 100
).round(2)

region = region.sort_values(
    'Total_Sales', ascending=False)

print(region.to_string(index=False))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Chart 1 - Sales by Region
axes[0].barh(region['Region'],
             region['Total_Sales'] / 1e6,
             color='#1F3864')
axes[0].set_title('Total Sales by Region ($M)',
                   fontsize=13, fontweight='bold',
                   color='#1F3864')
axes[0].set_xlabel('Sales ($M)')

# Chart 2 - Profit Margin by Region
colors = ['#2ECC71' if x > 10
          else '#E8722A' if x > 5
          else '#E74C3C'
          for x in region['Profit Margin']]

axes[1].barh(region['Region'],
             region['Profit Margin'],
             color=colors)
axes[1].set_title('Profit Margin % by Region',
                   fontsize=13, fontweight='bold',
                   color='#1F3864')
axes[1].set_xlabel('Margin %')
axes[1].axvline(x=0, color='black',
                linewidth=0.8, linestyle='--')

plt.tight_layout()
plt.savefig('regional_analysis.png',
            dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved !")
```

## 💸 Discount Impact Analysis

```python
# Sales and Profit by Discount Group
discount = orders.groupby(
    'Discount Group', observed=True).agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum'),
    Total_Orders=('Order ID', 'count'),
    Avg_Discount=('Discount', 'mean')
).reset_index()

discount['Profit Margin'] = (
    discount['Total_Profit'] / discount['Total_Sales'] * 100
).round(2)

print(discount.to_string(index=False))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Colors par groupe
colors = ['#2ECC71', '#E8722A', '#E74C3C', '#8E44AD']

# Chart 1 - Profit by Discount Group
axes[0].bar(discount['Discount Group'],
            discount['Total_Profit'] / 1e3,
            color=colors)
axes[0].set_title('Total Profit by Discount Group ($K)',
                   fontsize=13, fontweight='bold',
                   color='#1F3864')
axes[0].set_xlabel('')
axes[0].set_ylabel('Profit ($K)')
axes[0].axhline(y=0, color='black',
                linewidth=0.8, linestyle='--')

# Chart 2 - Profit Margin by Discount Group
axes[1].bar(discount['Discount Group'],
            discount['Profit Margin'],
            color=colors)
axes[1].set_title('Profit Margin % by Discount Group',
                   fontsize=13, fontweight='bold',
                   color='#1F3864')
axes[1].set_xlabel('')
axes[1].set_ylabel('Margin %')
axes[1].axhline(y=0, color='black',
                linewidth=0.8, linestyle='--')

plt.tight_layout()
plt.savefig('discount_analysis.png',
            dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved !")
```

## 📅 Time Trends Analysis

```python
# Sales and Profit by Year
yearly = orders.groupby('Year').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum'),
    Total_Orders=('Order ID', 'count')
).reset_index()

yearly['Profit Margin'] = (
    yearly['Total_Profit'] / yearly['Total_Sales'] * 100
).round(2)

print(yearly.to_string(index=False))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Chart 1 - Sales by Year
axes[0].plot(yearly['Year'],
             yearly['Total_Sales'] / 1e6,
             marker='o', color='#1F3864',
             linewidth=2.5, markersize=8)
axes[0].fill_between(yearly['Year'],
                     yearly['Total_Sales'] / 1e6,
                     alpha=0.3, color='#1F3864')
axes[0].set_title('Total Sales by Year ($M)',
                   fontsize=13, fontweight='bold',
                   color='#1F3864')
axes[0].set_xlabel('Year')
axes[0].set_ylabel('Sales ($M)')

# Chart 2 - Profit Margin by Year
axes[1].plot(yearly['Year'],
             yearly['Profit Margin'],
             marker='o', color='#2ECC71',
             linewidth=2.5, markersize=8)
axes[1].fill_between(yearly['Year'],
                     yearly['Profit Margin'],
                     alpha=0.3, color='#2ECC71')
axes[1].set_title('Profit Margin % by Year',
                   fontsize=13, fontweight='bold',
                   color='#1F3864')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('Margin %')

plt.tight_layout()
plt.savefig('time_trends.png',
            dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved !")
```

## 🏆 Top Products Analysis

```python
# Top 10 products by Sales
top_products = orders.groupby('Product Name').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum'),
    Total_Orders=('Order ID', 'count')
).reset_index()

top_products['Profit Margin'] = (
    top_products['Total_Profit'] /
    top_products['Total_Sales'] * 100
).round(2)

top_products = top_products.sort_values(
    'Total_Sales', ascending=False).head(10)

print(top_products[['Product Name',
                     'Total_Sales',
                     'Total_Profit',
                     'Profit Margin']].to_string(index=False))
```

```python
fig, ax = plt.subplots(figsize=(14, 6))

colors = ['#2ECC71' if x > 0
          else '#E74C3C'
          for x in top_products['Total_Profit']]

ax.barh(top_products['Product Name'],
        top_products['Total_Sales'] / 1e3,
        color='#1F3864')

ax.set_title('Top 10 Products by Total Sales ($K)',
              fontsize=13, fontweight='bold',
              color='#1F3864')
ax.set_xlabel('Sales ($K)')
ax.invert_yaxis()

plt.tight_layout()
plt.savefig('top_products.png',
            dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved !")
```

```python
# Bottom 10 products by Profit
bottom_products = orders.groupby('Product Name').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum')
).reset_index()

bottom_products['Profit Margin'] = (
    bottom_products['Total_Profit'] /
    bottom_products['Total_Sales'] * 100
).round(2)

bottom_products = bottom_products.sort_values(
    'Total_Profit', ascending=True).head(10)

print("Top 10 loss-making products :")
print(bottom_products[['Product Name',
                        'Total_Sales',
                        'Total_Profit',
                        'Profit Margin']].to_string(index=False))
```

## 👥 Customer Segment Analysis

```python
# Sales and Profit by Segment
segment = orders.groupby('Segment').agg(
    Total_Sales=('Sales', 'sum'),
    Total_Profit=('Profit', 'sum'),
    Total_Orders=('Order ID', 'count'),
    Avg_Order_Value=('Sales', 'mean')
).reset_index()

segment['Profit Margin'] = (
    segment['Total_Profit'] /
    segment['Total_Sales'] * 100
).round(2)

segment = segment.sort_values(
    'Total_Sales', ascending=False)

print(segment.to_string(index=False))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

colors = ['#1F3864', '#2980B9', '#95A5A6']

# Chart 1 - Sales by Segment
axes[0].bar(segment['Segment'],
            segment['Total_Sales'] / 1e6,
            color=colors)
axes[0].set_title('Total Sales by Segment ($M)',
                   fontsize=12, fontweight='bold',
                   color='#1F3864')
axes[0].set_ylabel('Sales ($M)')

# Chart 2 - Profit Margin by Segment
axes[1].bar(segment['Segment'],
            segment['Profit Margin'],
            color=colors)
axes[1].set_title('Profit Margin % by Segment',
                   fontsize=12, fontweight='bold',
                   color='#1F3864')
axes[1].set_ylabel('Margin %')

# Chart 3 - Average Order Value by Segment
axes[2].bar(segment['Segment'],
            segment['Avg_Order_Value'],
            color=colors)
axes[2].set_title('Average Order Value by Segment ($)',
                   fontsize=12, fontweight='bold',
                   color='#1F3864')
axes[2].set_ylabel('Avg Order Value ($)')

plt.tight_layout()
plt.savefig('segment_analysis.png',
            dpi=150, bbox_inches='tight')
plt.show()
print("Chart saved !")
```

## 🔗 Bivariate Analysis

### Quantitative vs Quantitative - Correlation Matrix

```python
# Select numerical columns
numerical = orders[[
    'Sales', 'Profit', 'Quantity',
    'Discount', 'Shipping Cost'
]]

# Correlation matrix
corr_matrix = numerical.corr().round(2)
print("Correlation Matrix :")
print(corr_matrix)

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix,
            annot=True,
            fmt='.2f',
            cmap='RdYlGn',
            center=0,
            square=True,
            linewidths=0.5)

plt.title('Correlation Matrix - Key Variables',
           fontsize=13, fontweight='bold',
           color='#1F3864')

plt.tight_layout()
plt.savefig('correlation_heatmap.png',
            dpi=150, bbox_inches='tight')
plt.show()
print("Heatmap saved !")
```

### Qualitative vs Qualitative - Cross Tables


```python
# Cross table Category vs Segment
cross_cat_seg = pd.crosstab(
    orders['Category'],
    orders['Segment'],
    values=orders['Sales'],
    aggfunc='sum'
).round(0)

print("Sales by Category and Segment ($) :")
print(cross_cat_seg)
print()

# Cross table Category vs Discount Group
cross_cat_disc = pd.crosstab(
    orders['Category'],
    orders['Discount Group'],
    values=orders['Profit Margin'],
    aggfunc='mean'
).round(3)

print("Average Profit Margin by Category and Discount Group :")
print(cross_cat_disc)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Chart 1 - Heatmap Category vs Segment
sns.heatmap(cross_cat_seg / 1e6,
            annot=True,
            fmt='.1f',
            cmap='Blues',
            ax=axes[0],
            linewidths=0.5)
axes[0].set_title('Sales ($M) by Category and Segment',
                   fontsize=12, fontweight='bold',
                   color='#1F3864')
axes[0].set_xlabel('Segment')
axes[0].set_ylabel('Category')

# Chart 2 - Heatmap Category vs Discount Group
sns.heatmap(cross_cat_disc,
            annot=True,
            fmt='.2f',
            cmap='RdYlGn',
            center=0,
            ax=axes[1],
            linewidths=0.5)
axes[1].set_title('Avg Profit Margin by Category and Discount Group',
                   fontsize=12, fontweight='bold',
                   color='#1F3864')
axes[1].set_xlabel('Discount Group')
axes[1].set_ylabel('Category')

plt.tight_layout()
plt.savefig('cross_tables.png',
            dpi=150, bbox_inches='tight')
plt.show()
print("Cross tables saved !")
```

### Quantitative - Mean Comparison by Group


```python
# Mean comparison by Category
print("=" * 55)
print("MEAN COMPARISON BY CATEGORY")
print("=" * 55)
mean_category = orders.groupby('Category').agg(
    Avg_Sales=('Sales', 'mean'),
    Avg_Profit=('Profit', 'mean'),
    Avg_Discount=('Discount', 'mean'),
    Avg_Margin=('Profit Margin', 'mean')
).round(3)
print(mean_category.to_string())

print("\n" + "=" * 55)
print("MEAN COMPARISON BY DISCOUNT GROUP")
print("=" * 55)
mean_discount = orders.groupby(
    'Discount Group', observed=True).agg(
    Avg_Sales=('Sales', 'mean'),
    Avg_Profit=('Profit', 'mean'),
    Avg_Margin=('Profit Margin', 'mean'),
    Count=('Order ID', 'count')
).round(3)
print(mean_discount.to_string())

print("\n" + "=" * 55)
print("MEAN COMPARISON BY SEGMENT")
print("=" * 55)
mean_segment = orders.groupby('Segment').agg(
    Avg_Sales=('Sales', 'mean'),
    Avg_Profit=('Profit', 'mean'),
    Avg_Margin=('Profit Margin', 'mean')
).round(3)
print(mean_segment.to_string())
```
